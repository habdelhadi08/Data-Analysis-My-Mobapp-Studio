```python
def load_dataset():
    df = pd.read_csv("googleplaystore.csv")
    return df
dataset=load_dataset()

```

```python
def print_summarize_dataset(dataset):
    print(dataset.head())
    print(dataset.info())
    print(dataset.shape)
    print(dataset.describe())
    print(dataset.isnull().sum())

print_summarize_dataset(dataset)
```

```python
def clean_dataset(dataset):
    dataset = dataset.dropna()
    dataset = dataset.drop_duplicates()
    dataset['Size'] = dataset['Size'].str.replace('M', '000').str.replace('k', '').str.replace('Varies with device', '0').astype(float)
    dataset['Price'] = dataset['Price'].str.replace('$', '').astype(float)
    return dataset
clean_dataset(dataset)
print(clean_dataset(dataset).head())
```

```python
def print_histograms(dataset):
    top10_categories = dataset['Category'].value_counts().head(30)
    plt.figure(figsize=(12, 10))
    plt.hist(top10_categories, color="blue", bins=40)
    plt.title('Top 10 categories')
    plt.xlabel("Category")
    plt.ylabel("frequency")
    plt.grid(True)
    plt.show()
print_histograms(dataset)
    
```

```python
def compute_correlations_matrix(dataset):
    df = pd.read_csv("googleplaystore.csv")
    matrix = df.corr()
    return matrix
compute_correlations_matrix(dataset)
```

```python
def print_scatter_matrix(dataset):
    plt.figure(figsize=(10,12))
    plt.scatter(dataset['Category'],dataset['Price'], color="blue")
    plt.title('Category vs. Price')
    plt.xlabel("Category")
    plt.ylabel("price")
    plt.grid(True)
    plt.show()
print_scatter_matrix(dataset)

```
